RENAMING COLUMN HEADER

```python
bank.rename(columns={
    'CLIENTNUM': 'cid',
    'Attrition_Flag': 'cstatus',
    'Customer_Age': 'cage',
    'Gender': 'gender',
    'Dependent_count': 'dcount',
    'Education_Level': 'education'
}, inplace=True)
```

DROPPING UNWANTED COLUMNS

```python
bank.drop([ 'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Avg_Open_To_Buy','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
       'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'], axis=1, inplace=True)
```

```python
bank.drop(['Total_Trans_Ct','Total_Amt_Chng_Q4_Q1','Total_Relationship_Count',],axis=1,inplace=True)
```

BI VARIANT ANALYSIS

```python
bank.columns
```

```python
sns.boxplot(data=bank, x='gender', y='Total_Trans_Amt')
plt.title("Average Transaction Amount by Gender")
plt.show()
```

```python
sns.countplot(data=bank, x='education', hue='Income_Category')
plt.title("Education Level vs Income")
plt.xticks(rotation=45)
plt.legend(title='Income')
plt.show()
```

```python
sns.countplot(data=bank, x='gender', hue='Marital_Status')
plt.title("Gender vs Marital Status")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=bank, x='Income_Category', y='Credit_Limit', palette='Set2')

plt.title("Credit Limit by Income Category")
plt.xlabel("Income Category")
plt.ylabel("Credit Limit")
```

```python
avg_util = bank.groupby('Income_Category')['Avg_Utilization_Ratio'].mean().sort_values(ascending=False)
avg_util.plot(kind='bar', color='teal', figsize=(7,4))
plt.ylabel("Average Utilization Ratio")
plt.title("Avg Credit Utilization by Income Category")
```

```python
bank.groupby('Marital_Status')['Total_Trans_Amt'].mean().plot(kind='bar')
```

```python
sns.scatterplot(data=bank, x='cage', y='Total_Trans_Amt', hue='gender')
plt.title("Age vs Transaction Amount")
```

```python
sns.scatterplot(data=bank, x='Credit_Limit', y='Total_Trans_Amt', hue='gender')
plt.title("Total Transaction Amount vs Credit Limit")
plt.show()
```

```python
util_by_edu = bank.groupby('education')['Avg_Utilization_Ratio'].mean().sort_values()
util_by_edu.plot(kind='barh', color='coral')
plt.title("Avg Credit Utilization by Education Level")
plt.xlabel("Avg Utilization")
```
